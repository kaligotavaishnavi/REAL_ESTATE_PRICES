# IMPLEMENTATING LINEAR REGRESSION MODEL

# STEP 1: IMPORTING LIBRARIES 

```python
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import warnings
warnings.simplefilter('ignore')
#plt.style.use('dark_background')
```

```python
data= pd.read_csv('Transformed_Housing_Data2.csv')
```

```python
data.head()
```

# STEP 2: SCALING THE DATASET

```python
from sklearn.preprocessing import StandardScaler
scaler= StandardScaler()
Y=data['Sale_Price']
X=scaler.fit_transform(data.drop(columns=['Sale_Price']))
X=pd.DataFrame(data=X,columns=data.drop(columns=['Sale_Price']).columns)
X.head()
```

# STEP 3: CHECKING AND REMOVING MULTICOLLINEARTY

```python
X.corr()
```

```python
## pair of independent variables with correlation greater than 0.5
k= X.corr()
z=[[str(i),str(j)]for i in k.columns for j in k.columns if(k.loc[i,j]>abs(0.5))&(i!=j)]
z,len(z)
```

```python
#importing variancve_inflation_factor function from the statsmodels
from statsmodels.stats.outliers_influence import variance_inflation_factor
vif_data = X
#calculating VIF for every column
VIF= pd.Series([variance_inflation_factor(vif_data.values,i)for i in range(vif_data.shape[1])],index=vif_data.columns)
VIF
```

```python
VIF[VIF==VIF.max()].index[0]
```

```python
def MC_remover(data):
    vif=pd.Series([variance_inflation_factor(vif_data.values,i)for i in range(vif_data.shape[1])],index=vif_data.columns)
    if vif.max()>5:
        print(vif[vif==vif.max()].index[0],'has been removed')
        data=data.drop(columns=[vif[vif==vif.max()].index[0]])
        return data
    else:
        print('no multicollinearity anymore')
        return data
```

```python
for i in range(7):
    vif_data=MC_remover(vif_data)
    vif_data.head()
```

```python
#calculating VIF for remaining columns
VIF= pd.Series([variance_inflation_factor(vif_data.values,i)for i in range(vif_data.shape[1])],index=vif_data.columns)
VIF,len(vif_data.columns)
```

# STEP 4: TRAINING AND TESTING DATA

```python
X=vif_data
Y=data['Sale_Price']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(X,Y,test_size=0.3,random_state=101)
x_train.shape,x_test.shape,y_train.shape,y_test.shape
```

# STEP 5: LINEAR REGRESSION

```python
from sklearn.linear_model import LinearRegression
lr= LinearRegression(normalize=True)
lr.fit(x_train,y_train)
```

```python
lr.coef_
```

# STEP 6: GENERATING PREDICTIONS


```python
predictions=lr.predict(x_test)
lr.score(x_test,y_test)
```

# STEP 8: PLOTTING RESIDUALS

```python
residuals=predictions-y_test
residual_table= pd.DataFrame({'residuals':residuals,
                             'predictions':predictions})
residual_table=residual_table.sort_values(by='predictions')
```

```python
z=[i for i in range(int(residual_table['predictions'].max()))]
k=[0 for i in range(int(residual_table['predictions'].max()))]
```

```python
plt.figure(dpi=130,figsize=(17,7))
plt.scatter(residual_table['predictions'],residual_table['residuals'],color='red',s=2)
plt.plot(z,k,color='green',linewidth=3,label='regression line')
plt.ylim(-800000,800000)
plt.xlabel('fitted points(ordered by predictions)')
plt.ylabel('residuals')
plt.title('residual plot')
plt.legend()
plt.show()
```

# STEP 9: VERIFYING THE ASSUMPTIONS OF LINEAR REGRESSION

```python
plt.figure(dpi=130,figsize=(10,7))
plt.hist(residual_table['residuals'],color='red',bins=200)
plt.plot(z,k,color='green',linewidth=3,label='regression line')
plt.xlabel('residuals')
plt.ylabel('frequency')
plt.title('distribution of errors')
plt.show()
```

# INTERPRETING THE RESULTS OF LINEAR REGRESSION

MODEL COEFFICIENTS

```python
coefficient_table= pd.DataFrame({'column':x_train.columns,
                                'coefficients':lr.coef_})
coefficient_table=coefficient_table.sort_values(by='coefficients')
```

```python
plt.figure(dpi=130,figsize=(9,7))
x=coefficient_table['column']
y=coefficient_table['coefficients']
plt.barh(x,y)
plt.xlabel('coefficient')
plt.ylabel('variables')
plt.title('normalized coeff')
plt.show()
```

# COMBINATION OF FEATURES 

```python
data['Ever Renovate']=np.where(data['Renovated Year']==0,'No','Yes')
```

```python
data['Purchase Year']= pd.DatetimeIndex(data['Date House was Sold']).year
data['Year Since Renovation']=np.where(data['Ever Renovate']=='Yes',abs(data['Purchase Year']-data['Renovated Year']),0)
```

```python
def luxury_home(row):
    count=0
    if row[0]=='Yes':
        count=count+1
    if row[1] in ['Good','Excellent']:
        count=count+1
    if row[2]>=8:
        count=count+1
    if count>=2:
        return 'Yes'
    else:
        return 'No'
    
data['luxury_home']=data[['Waterfront View','Condition of the House','Overall Grade']].apply(luxury_home,axis=1)
data['luxury_home'].unique()
```

```python
luxury=data.pivot_table(values='Sale Price',columns='luxury_home',aggfunc='mean')
luxury
```

```python
plt.figure(dpi=80,figsize=(5,8))
plt.bar(luxury.columns,luxury.iloc[0])
plt.xlabel('luxury home')
plt.ylabel('mean sale price')
plt.show()
```
